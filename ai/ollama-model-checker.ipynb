````python
# Function to check if Ollama is installed
def check_ollama_installed():
    try:
        result = subprocess.run(['which', 'ollama'], capture_output=True, text=True)
        if result.returncode == 0:
            return True
        else:
            return False
    except Exception as e:
        return False
````

````python
# Function to check if a model is available for pulling
def check_model_availability(model_name):
    try:
        # Use timeout to prevent hanging
        result = subprocess.run(
            ['timeout', '30', 'ollama', 'pull', model_name],
            capture_output=True,
            text=True,
            timeout=35
        )

        if result.returncode == 0:
            return True, "Available"
        else:
            return False, result.stderr if result.stderr else "Not available"
    except subprocess.TimeoutExpired:
        return False, "Timeout - Model download took too long"
    except Exception as e:
        return False, f"Error: {str(e)}"
````

````python
# Function to get currently installed models
def get_installed_models():
    try:
        result = subprocess.run(['ollama', 'list'], capture_output=True, text=True)
        if result.returncode == 0:
            return result.stdout
        else:
            return "Error retrieving models"
    except Exception as e:
        return f"Error: {str(e)}"
````

````python
# Check Ollama installation
display(Markdown("# Ollama Model Availability Checker"))
display(Markdown("## Checking Ollama Installation"))

if check_ollama_installed():
    display(Markdown("✓ Ollama is installed and accessible"))
else:
    display(Markdown("✗ Ollama is not installed or not in PATH"))
````

````python
# Display model availability check results
display(Markdown("## Model Availability Check Results"))

# Get current models before checking
display(Markdown("### Currently Installed Models"))
current_models = get_installed_models()
display(Markdown(f"```{current_models}```"))
````

````python
# Check each model individually
display(Markdown("### Checking Individual Models"))
results = []

for i, model in enumerate(models_to_check):
    is_available, message = check_model_availability(model)
    results.append((model, is_available, message))

    if is_available:
        status = "✓ Available"
    else:
        status = "✗ Not Available"

    display(Markdown(f"**{i+1}. {model}**: {status}"))

    if not is_available and message != "Timeout - Model download took too long":
        display(Markdown(f"  Error details: {message}"))
````

````python
# Summary statistics
display(Markdown("## Summary"))
available_count = sum(1 for _, available, _ in results if available)
total_count = len(results)

display(Markdown(f"**Total Models Checked**: {total_count}"))
display(Markdown(f"**Available Models**: {available_count}"))
display(Markdown(f"**Not Available Models**: {total_count - available_count}"))
````

````python
# Installation Commands
display(Markdown("## Installation Commands"))
display(Markdown("### For Models That Are Not Available"))
for model, is_available, _ in results:
    if not is_available:
        display(Markdown(f"`ollama pull {model}`"))
````

````python
import subprocess, json, re, yaml
from pathlib import Path
from IPython.display import display, Markdown

# ── locate configs dir relative to this notebook ──────────────────────────────
NOTEBOOK_DIR = Path().resolve()
CONFIGS_DIR  = NOTEBOOK_DIR.parent / "scripts" / "configs"

# ── helpers ───────────────────────────────────────────────────────────────────

def strip_jsonc(text: str) -> str:
    """Remove // and /* */ comments from JSONC, leaving string contents intact."""
    def replacer(m):
        return " " if m.group(0).startswith("/") else m.group(0)
    return re.sub(r'("(?:[^"\\]|\\.)*")|(/\*.*?\*/|//[^\n]*)', replacer, text, flags=re.DOTALL)

def norm(model_id: str) -> str:
    """Canonical form for comparison: strip provider prefix, lowercase."""
    s = re.sub(r"^ollama/", "", model_id.strip())
    return s.lower()

def get_installed() -> dict[str, str]:
    """Return {normalized_name: 'NAME SIZE'} from `ollama list`."""
    out = subprocess.check_output(["ollama", "list"], text=True)
    installed: dict[str, str] = {}
    for line in out.splitlines()[1:]:           # skip header
        parts = line.split()
        if not parts:
            continue
        name = parts[0]
        size = f"{parts[2]} {parts[3]}" if len(parts) >= 4 else ""
        installed[norm(name)] = f"{name}  ({size})"
    return installed

# ── extract model IDs from every config file ──────────────────────────────────

def extract_models_opencode(path: Path) -> list[str]:
    data   = json.loads(strip_jsonc(path.read_text()))
    models = set()
    # catalog keys
    for mid in data.get("provider", {}).get("ollama", {}).get("models", {}).keys():
        models.add(mid)
    # top-level model / small_model
    for field in ("model", "small_model"):
        if v := data.get(field):
            models.add(re.sub(r"^ollama/", "", v))
    # agent models
    for agent in data.get("agent", {}).values():
        if v := agent.get("model"):
            models.add(re.sub(r"^ollama/", "", v))
    return list(models)

def extract_models_yaml(path: Path) -> list[str]:
    data = yaml.safe_load(path.read_text())
    return [m["model"] for m in (data.get("models") or []) if m.get("model")]

def extract_models_crush(path: Path) -> list[str]:
    data   = json.loads(path.read_text())
    models = set()
    for provider in data.get("providers", {}).values():
        for m in provider.get("models", []):
            if mid := m.get("id"):
                models.add(mid)
    for agent in data.get("agents", {}).values():
        if mid := agent.get("model"):
            models.add(mid)
    return list(models)

def extract_models_simple_json(path: Path) -> list[str]:
    data = json.loads(path.read_text())
    return list(data.get("provider", {}).get("ollama", {}).get("models", {}).keys())

EXTRACTORS = {
    "opencode.jsonc":      extract_models_opencode,
    "continue_config.yaml": extract_models_yaml,
    "crush.json":          extract_models_crush,
    "claude_code.yaml":    extract_models_yaml,
    "grok.json":           extract_models_simple_json,
    "gemini.json":         extract_models_simple_json,
}

# ── collect all referenced models ─────────────────────────────────────────────

referenced_raw: dict[str, list[str]] = {}   # norm_id -> [source files]
for fname, extractor in EXTRACTORS.items():
    path = CONFIGS_DIR / fname
    if not path.exists():
        continue
    for mid in extractor(path):
        key = norm(mid)
        referenced_raw.setdefault(key, [])
        if fname not in referenced_raw[key]:
            referenced_raw[key].append(fname)

# ── compare ───────────────────────────────────────────────────────────────────

installed    = get_installed()                         # norm -> display str
installed_k  = set(installed.keys())
referenced_k = set(referenced_raw.keys())

to_remove  = sorted(installed_k  - referenced_k)      # installed, not in any config
to_pull    = sorted(referenced_k - installed_k)        # in config, not installed
in_sync    = sorted(installed_k  & referenced_k)       # all good

# ── display ───────────────────────────────────────────────────────────────────

display(Markdown("### ✅ In sync — installed & referenced"))
if in_sync:
    rows = "\n".join(f"- `{installed[k]}`  ← {', '.join(referenced_raw[k])}" for k in in_sync)
    display(Markdown(rows))
else:
    display(Markdown("_none_"))

display(Markdown("---\n### ⬇️  Need to pull — in config but not installed"))
if to_pull:
    rows = "\n".join(f"- `{k}`  ← {', '.join(referenced_raw[k])}" for k in to_pull)
    display(Markdown(rows))
    pull_cmds = "\n".join(f"ollama pull {k}" for k in to_pull)
    display(Markdown(f"```shell\n{pull_cmds}\n```"))
else:
    display(Markdown("_nothing to pull_"))

display(Markdown("---\n### 🗑️  Safe to remove — installed but not in any config"))
if to_remove:
    rows = "\n".join(f"- `{installed[k]}`" for k in to_remove)
    display(Markdown(rows))
    rm_cmds = "\n".join(f'ollama rm "{k}"' for k in to_remove)
    display(Markdown(f"```shell\n{rm_cmds}\n```"))
else:
    display(Markdown("_nothing to remove — configs and Ollama are in sync_ 🎉"))
````
